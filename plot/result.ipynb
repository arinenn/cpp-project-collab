# Fast Fourier Transform in Heston model

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')

K_lower = 0.65
K_upper = 1.35
df = pd.read_csv("output.csv")
ind = (df["K"] >= K_lower) & (df["K"] <= K_upper)
plt.figure(figsize=(9, 6))

plt.title("Call Options Prices in Heston model")
plt.plot(df["K"][ind], df["C"][ind], color = 'r', linewidth = 3, label = 'T = 0.5 year')
plt.xlabel("K, strikes")
plt.ylabel("C, prices")
plt.legend()
plt.show()
```

```python
data = pd.read_csv("input.csv")
ind = (df["K"] >= K_lower) & (df["K"] <= K_upper)
plt.figure(figsize=(9, 6))

plt.title("Error")
plt.plot(df["K"][ind], df["C"][ind].values - data["C"].values, color = 'r', linewidth = 3, label = 'T = 0.5 year')
plt.xlabel("K, strikes")
plt.ylabel("errors")
plt.legend()
plt.show()
```
