# Fast Fourier Transform in Heston model

Here you can look through calculated data and make plots for next examples:
- ```example-1``` (call options $T=0.5, K\in[0.65,1.35]$ for $S_0=1$)
- ```example-2``` (put options $T=1, K\in[65,135]$ for $S_0=100$)

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')
plt.rcParams['figure.dpi'] = 300
````

````python
# Gather data
df = pd.read_csv("example-1.csv")

# Make figure
plt.figure(figsize=(9, 6))
plt.title("Call Options Prices in Heston model")
plt.plot(df["K"], df["C"], color = 'r', linewidth = 3, label = 'T = 0.5 year')
plt.xlabel("K, strikes")
plt.ylabel("C, prices")
plt.legend()

# Save figure
plt.savefig('example-1.png')
````

````python
# Gather data
df = pd.read_csv("example-2.csv")

# Make figure
plt.figure(figsize=(9, 6))
plt.title("Put Options Prices in Heston model")
plt.plot(df["K"], df["C"], color = 'r', linewidth = 3, label = 'T = 1 year')
plt.xlabel("K, strikes")
plt.ylabel("C, prices")
plt.legend()

# Save figure
plt.savefig('example-2.png')
````
